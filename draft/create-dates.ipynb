```python
import requests
import pandas as pd
from datetime import date
from bs4 import BeautifulSoup
```

```python
pd.date_range(start = '2020-02-07', end = date.today(), freq = 'D')
```

```python
df = pd.DataFrame(
    {'date': pd.date_range(start = '2020-02-07', end = date.today(), freq = 'D')}
)
df
```

```python
df['url_short'] = 'moh' + df['date'].dt.strftime('%-d%b').str.lower()
df
```

```python
df['url_short_ext'] = 'https://go.gov.sg/' + df['url_short']
df
```

```python
def get_long_url(url_short):
    url_long = ''
    header = {
        'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:77.0) Gecko/20100101 Firefox/77.0'
    }
    
    response = requests.get(url_short, headers = header)
    
    if(response.status_code == 200):
        content = response.content if response.status_code == 200 else ''
        soup = BeautifulSoup(content, 'html.parser')
        url_long = soup.find('p').get('data-href')
    
    return url_long

get_long_url('https://go.gov.sg/moh24apr')
```

```python
df['url_long'] = df['url_short_ext'].apply(lambda url: get_long_url(url))
df
```

```python
df.to_csv('out/urls.csv')
```
