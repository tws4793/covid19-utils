```python
import pandas as pd
import numpy as np
import datetime
import string
import re
```

```python
def preprocess(df, year = 2020):
    df.columns = ['date', 'time', 'location_address']
    
    # Raw
    df['date'] = df['date'] + ' ' + str(year)
    df['raw_time_start'] = df['date'] + ' ' + df['time'].str[0:4]
    df['raw_time_end'] = df['date'] + ' ' + df['time'].str[-5:-1]
    df['end_new_day'] = df['raw_time_end'] < df['raw_time_start']

    # Processed
    df['time_start'] = pd.to_datetime(df['raw_time_start'], format = '%d %b %Y %H%M')
    df['time_end'] = pd.to_datetime(df['raw_time_end'], format = '%d %b %Y %H%M')
    df['name'] = df['location_address'].str.split('(').str[0]
    df['address'] = df['location_address'].str.extract('.*\((.*)\).*')
    df['raw_sublocations'] = df['location_address'].str.split(')').str[1]
    
    # Delete
    df = df.drop(['end_new_day'], axis = 1)
#     df['sublocations'] = df['raw_sublocations'].apply(get_sub_location)
    
    return df

def get_sub_location(places, delimiters = None, is_str = True):
    if delimiters == None:
        delimiters = ['\u2022', '\uf0b7']
    
    if places == None | places == '':
        return []
        
    splitting = re.compile('|'.join(delimiters))
    
    places_list = splitting.split(places)
    places_list = [s.strip() for s in places_list]
    places_list = [s for s in places_list if s]
    
    return places_list
```
